## Program 1: Swap Two Numbers

## Program 14: Prime Check

**Problem**: Read an integer `n` and print `Prime` or `Not Prime`.

**Sample Input**:
```
29
```
**Sample Output**:
```

```


````python
"""
This program reads an integer n and print Prime or Not Prime.
Handles Invalid Input with try/except
"""

try:
    # Take  input, convert  to int, assign  to n
    n = int(input("Input an Integer: "))  # `int` ensures only int is taken here

    if n < 2: # To handle 1 which is not a rime number
        print("Not Prime")

    else:

        prime_val = 1   # initializing
        for num in range(2, n):  # starting from 2

            if n % num == 0:
                prime_val = 0
                # print("Not Prime")
                break

        if prime_val == 1:
            print("Prime")
        else:
            print("Not Prime")


except ValueError as e:

    print("Invalid Input! Pls input Integer Only! Error -->: ", e)
````

## Program 28: Second Largest (Distinct)

**Problem**: Read `N` and a list of integers. Print the second largest **distinct** number.

**If second largest does not exist**, print: `Not Possible`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Second largest distinct value or `Not Possible`

**Sample Input**:
```
5
5 1 5 3 2
```
**Sample Output**:
```
3
```

## Program 30: List Palindrome Check

**Problem**: Read `N` and a list of `N` integers. Print `Palindrome` if the list reads the same forward and backward, else print `Not Palindrome`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- `Palindrome` or `Not Palindrome`

**Sample Input**:
```
5
1 2 3 2 1
```
**Sample Output**:
```
Palindrome
```


````python
"""
This program reads N and a list of N integers. Print Palindrome if the list reads the same forward and backward, else print Not Palindrome
Handles Invalid Input with try/except
"""

try:
    # Take  input string
    N = int(input("Input Number of Integers N in your List : "))  # take N integer
    input_integers = map(
        int,
        (input(f"Input the {N} Integers  separated by space: ")).split(),
    )  # input elements

    integer_list = list(input_integers)  # convert to list

    if len(integer_list) == N:  # check number of integers

        integer_list_r = integer_list[::-1]  # reverse the list

        if integer_list == integer_list_r:  # compare original and reversed lists
            print("Palindrome")

        else:
            print("Not Palindrome")

    else:
        print(f"You must input  exactly {N} integers space-separated")


except ValueError as e:

    print("Invalid Input! Pls input integer Only: Error -->: ", e)
````
